```python
import csv
import random
user_set = set()

def fix_nulls(s):
    for line in s:
        yield line.replace('\0', ' ')

with open("", encoding="utf-8") as csvfile:
    reader = csv.reader(fix_nulls(csvfile))
    next(reader, None) # skip header
    for line in reader:
        user_set.add(line[0])

sample = random.sample(list(user_set), 10000)
```

```python
import csv
import pickle
from collections import defaultdict

sample_set = set(sample)
sample_map = {sample[i]:str(i) for i in range(len(sample))}

print(len(sample), len(sample_set), len(sample_map))

users = defaultdict(dict)
pf_view = set()
user_must = set()

with open("", encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    next(reader, None)
    for line in reader:
        if line[1] in sample_set and line[2] in sample_set:
            user_must.add(line[1])
            user_must.add(line[2])


for user in user_must:
    sample_set.remove(user)
resample = random.sample(list(sample_set), 1000 - len(user_must))
resample_set = set(resample)

print(len(resample_set), len(user_must))
new_sample_set = resample_set | user_must
new_sample = list(new_sample_set)
print(len(new_sample))

new_sample_map = {new_sample[i]:str(i) for i in range(len(new_sample))}

with open("", encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    next(reader, None)
    for line in reader:
        if line[1] in new_sample_set and line[2] in new_sample_set:
            if line[3] == "pf_view":
                users[new_sample_map[line[1]]][new_sample_map[line[2]]] = 5 if users[new_sample_map[line[1]]].get(new_sample_map[line[2]], 0) + 2.5 > 5 else users[new_sample_map[line[1]]].get(line[2], 0) + 2.5
            elif line[3] == "mem_concn":
                users[new_sample_map[line[1]]][new_sample_map[line[2]]] = 5 if users[new_sample_map[line[1]]].get(new_sample_map[line[2]], 0) + 2.5 > 5 else users[new_sample_map[line[1]]].get(line[2], 0) + 2.5

print(len(users))
print(users)

with open("user_rating.pkl", "wb") as f:
    pickle.dump(users, f)
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Flatten, ZeroPadding2D, Convolution2D, MaxPooling2D, Dropout, Activation

model = Sequential()
model.add(ZeroPadding2D((1,1),input_shape=(224,224, 3)))
model.add(Convolution2D(64, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
 
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(128, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
 
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(256, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
 
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
 
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Convolution2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
 
model.add(Convolution2D(4096, (7, 7), activation='relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(4096, (1, 1), activation='relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(2622, (1, 1)))
model.add(Flatten())
model.add(Activation('softmax'))

model.load_weights('./vgg_face_weights.h5')

vgg_face_descriptor = Model(inputs=model.layers[0].input, outputs=model.layers[-2].output)
```

```python
def load(image_file):
    # Read and decode an image file to a uint8 tensor
    image = tf.io.read_file(image_file)
    image = tf.io.decode_jpeg(image)

    input_image = tf.cast(image, tf.float32)
    
    return input_image

def resize(input_image, height, width):
    input_image = tf.image.resize(
        input_image,
        [height, width],
        method=tf.image.ResizeMethod.NEAREST_NEIGHBOR
    )
    
    return input_image

def normalize(input_image):
    input_image = (input_image / 127.5) - 1
    
    return input_image

def load_image(image_file):
    id = tf.strings.split(image_file, "/")[-1]
    id = tf.strings.split(id, ".")[0]
    input_image = load(image_file)
    input_image = resize(input_image, 224, 224)
    input_image = normalize(input_image)
    
    return input_image, id
```

```python
tf.config.set_visible_devices([], 'MPS')
```

```python
import tensorflow as tf
PATH = ""
BATCH_SIZE = 256

img_dataset = tf.data.Dataset.list_files(str(PATH + "*.jpg"))
img_dataset = img_dataset.map(
    load_image,
    num_parallel_calls=tf.data.AUTOTUNE
)
img_dataset = img_dataset.batch(BATCH_SIZE)

image_feature_vector = [[vgg_face_descriptor(input), id] for input, id in img_dataset]
```

```python
# Fit IncrementalPCA on each of subset
from sklearn.decomposition import IncrementalPCA

n = len(image_feature_vector)

ipca = IncrementalPCA(n_components=100, batch_size=BATCH_SIZE)

for i in range(n):
    if len(image_feature_vector[i][0]) >= 100:
        ipca.partial_fit(image_feature_vector[i][0])
    
```

```python
import numpy as np
from collections import defaultdict
from sklearn.neighbors import NearestNeighbors

nn_model = []
user_profile_feature = {}

# Num of batches
for i in range(n):
     curr_batch = image_feature_vector[i]
     feature_vec_len = len(curr_batch[0])
     index_vec_len = len(curr_batch[1])
     
     assert feature_vec_len == index_vec_len, "Wrong feature shapes"

     pca_res = ipca.transform(curr_batch[0])
     n_neighbors = min(feature_vec_len, 100)
     
     neighbor_nn = NearestNeighbors(n_neighbors=n_neighbors, radius=0.5, algorithm='brute', metric='cosine')
     neighbor_nn.fit(pca_res)
     idx_list = list(map(lambda x : x.decode('UTF-8'), curr_batch[1].numpy().tolist()))
     nn_model.append([neighbor_nn, idx_list])
     
     user_profile_feature.update({idx:feature for feature, idx in (zip(pca_res, idx_list))})
```

```python
print(user_profile_feature)
```

```python
# Calculate similarity
# total size of user_similarity = num of batches * num of nn
user_similarity_res = {}
for user in user_profile_feature.keys():
    tmp = []
    for nn, idx_list in nn_model:
        dist, ind = nn.kneighbors([user_profile_feature[user]])
        tmp.extend([(dist[0][i], idx_list[ind[0][i]]) for i in range(len(dist[0]))])
    user_similarity_res[user] = sorted(tmp, key = lambda x : x[0])[:100]
```
